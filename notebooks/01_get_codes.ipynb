```python
import os
os.environ.setdefault("GCLOUD_PROJECT", "enduring-rush-410922")

from google.cloud import bigquery
client = bigquery.Client()
```

```python

# Perform a query.
QUERY = '''
SELECT series_code, indicator_name
FROM bigquery-public-data.world_bank_wdi.series_summary
'''

query_job = client.query(QUERY)  # API request
df = query_job.to_dataframe()
```

```python
df.to_csv('../data/codes.csv')
df
```

```python
QUERY = """
SELECT country_code, 
short_name Name,
region Region,
income_group `Income Group`
FROM bigquery-public-data.world_bank_wdi.country_summary
ORDER BY country_code
"""

query_job = client.query(QUERY)  # API request
df = query_job.to_dataframe()
```

```python
df.to_csv('../data/countries.csv')
df
```
